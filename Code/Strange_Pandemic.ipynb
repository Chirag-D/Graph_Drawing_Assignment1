Import all necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read CSV

```python
path = '/content/AH_Excess_Deaths_by_Sex__Age__and_Race_and_Hispanic_Origin.csv'
data = pd.read_csv(path)
data.head()
```

Plot the graph

```python
# Filter the data for weeks 1-34 and years 2018-2022
filtered_data = filtered_data = data[(data['Geography'] == 'United States') & (data['MMWRweek'] >= 1) & (data['MMWRweek'] <= 34) & (data['MMWRyear'] >= 2018) & (data['MMWRyear'] <= 2022)]

# Group the data by year and sum the deaths for each year
annual_deaths = filtered_data.groupby('MMWRyear')['Deaths (unweighted)'].sum().reset_index()

# Plot the summed data as a horizontal bar chart
plt.figure(figsize=(16, 4))
plt.barh(annual_deaths['MMWRyear'].astype(str), annual_deaths['Deaths (unweighted)'])
plt.xlabel('Total Deaths')
plt.ylabel('Year')
plt.title('Strange Pandemic: MORE Death in Ending Years', fontsize=16, color='red', fontweight='bold')
plt.grid(axis='x')
for index, value in enumerate(annual_deaths['Deaths (unweighted)']):
    plt.text(0.90 * value, index, str(value))
plt.show()
```
